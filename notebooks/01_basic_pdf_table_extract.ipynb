### Basic Text/Table Extraction

```python
import fitz  # PyMuPDF
import re
```

```python
pdf_path = "GP_Delimitation_2019_Final.pdf"
```

```python


def extract_text_from_pdf(pdf_path):
    text = ''
    pdf_document = fitz.open(pdf_path)
    
    for page_num in range(pdf_document.page_count):
        page = pdf_document[page_num]
        text += page.get_text("text")
    
    pdf_document.close()
    return text
```

```python
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
def extract_tables_from_text(text):
    # Split the text into lines
    lines = text.split('\n')
    
    # Initialize a list to store the extracted tables
    tables = []
    current_table = []
    
    # Iterate through the lines and identify tables
    for line in lines:
        # Check if the line contains spaces or tabs (potential table row)
        if re.search(r'\s+', line):
            row = re.split(r'\s+', line.strip())
            current_table.append(row)
        else:
            # If the current_table is not empty, add it to the tables list
            if current_table:
                tables.append(current_table)
                current_table = []
    
    return tables
```

```python
extracted_tables = extract_tables_from_text(pdf_text)
```

```python
# Print the extracted tables
for table in extracted_tables:
    for row in table:
        print(row)
    print("-" * 20)
```
